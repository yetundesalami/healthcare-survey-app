```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and clean the data
df = pd.read_csv("data.csv")
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Total Income"] = pd.to_numeric(df["Total Income"], errors="coerce")
df["Amount"] = pd.to_numeric(df["Amount"], errors="coerce")
df.dropna(subset=["Age", "Total Income", "Amount"], inplace=True)

# Visualization 1: Top 10 Ages with Highest Income
income_by_age = df.groupby("Age")["Total Income"].max().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
income_by_age.plot(kind="bar")
plt.title("Top 10 Ages with Highest Income")
plt.ylabel("Total Income")
plt.xlabel("Age")
plt.tight_layout()
plt.savefig("top_incomes_by_age.png")
plt.show()

# Visualization 2: Gender distribution across spending categories
pivot = df.pivot_table(index="Category", columns="Gender", values="Amount", aggfunc="sum")

pivot.plot(kind="bar", figsize=(10, 6))
plt.title("Spending by Gender Across Categories")
plt.ylabel("Amount Spent")
plt.xlabel("Expense Category")
plt.tight_layout()
plt.savefig("gender_spending.png")
plt.show()
```
